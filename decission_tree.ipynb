```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
# % humidity values
humidity = np.array([[40], [55], [60] , [65] ,[70] , [80] , [90]])
```

```python
# 0 = no rain , 1 = rain
rain = np.array([0,0,0,1,1,1,1])
```

```python
# train a Decision tree
model = DecisionTreeClassifier (max_depth=2).fit(humidity,rain)
```

```python
plt.figure(figsize=(12,8))
plot_tree(model,feature_names=['humidity'],class_names=['no rain', 'rain'])
plt.show()
```
